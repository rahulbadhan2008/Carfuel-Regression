```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cfl['gas_type']=le.fit_transform(cfl['gas_type'])
```

```python
cfl['distance']=cfl['distance'].astype(int)
cfl['consume']=cfl['consume'].astype(int)
cfl['temp_inside']=cfl['temp_inside'].astype(int)
```

## Decision-making


```python
x=cfl.drop(['consume','AC'],axis=1)

y=cfl['consume']

x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.40,random_state=43)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
'''from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
x_train=mms.fit_transform(x_train)
x_test= mms.transform(x_test)'''
```

```python
from sklearn.decomposition import PCA
pc=PCA(n_components=7)
x_train=pc.fit_transform(x_train)
x_test=pc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(criterion='mse',splitter='best',max_depth=3,random_state=72)
dt.fit(x_train,y_train)
py=dt.predict(x_test)
```

```python
from sklearn import metrics
import numpy as np
from sklearn.metrics import mean_squared_error
from math import sqrt
print('Root Mean Squared Error(RMSE):', np.sqrt(metrics.mean_squared_error(y_test, py)))
print('Mean Absolute Error(MAE):', metrics.mean_absolute_error(y_test, py))
print('Mean Squared Error(MSE):', metrics.mean_squared_error(y_test, py))
```

```python
plt.scatter(y_test,py)
plt.xlabel('y_test(True Values)')
plt.ylabel('py')
plt.show()
```

```python
sns.distplot((y_test-py),bins=50)
plt.show()
```

```python
ftr=['distance','speed','gas_type','rain','sun','temp_outside','temp_inside']

from sklearn.tree import export_graphviz
export_graphviz(dt,out_file="tree.dot",feature_names=ftr,rounded=True,filled=True)
from subprocess import call
call(['dot','-Tpng','tree.dot','-o','tree.png','-Gdpi=600'])

from IPython.display import Image
Image(filename='tree.png')
```

```python
from sklearn.ensemble import RandomForestRegressor
md=RandomForestRegressor(n_estimators='warn',criterion='mse', random_state=44, max_depth=3)
md.fit(x_train,y_train)
pyy=md.predict(x_test)
```

```python
print('Root Mean Squared Error(RMSE):', np.sqrt(metrics.mean_squared_error(y_test, pyy)))
print('Mean Absolute Error(MAE):', metrics.mean_absolute_error(y_test, pyy))
print('Mean Squared Error(MSE):', metrics.mean_squared_error(y_test, pyy))
```

```python
plt.scatter(y_test,pyy)
plt.xlabel('y_test(True Values)')
plt.ylabel('pyy')
plt.show()
```

```python
sns.distplot((y_test-pyy),bins=50)
plt.show()
```

```python
tree = md.estimators_[5]
from sklearn.tree import export_graphviz
import pydot
export_graphviz(tree, out_file = 'tree.dot', feature_names = ftr, rounded = True, precision = 1,filled=True)
(graph,) = pydot.graph_from_dot_file('tree.dot')
graph.write_png('tree2.png')
import matplotlib.image as mpimg
plt.imshow(mpimg.imread('tree2.png'))
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lnr=LinearRegression()
lnr.fit(x_train,y_train)
p_y=lnr.predict(x_test)
```

```python
print('Root Mean Squared Error(RMSE):', np.sqrt(metrics.mean_squared_error(y_test, p_y)))
print('Mean Absolute Error(MAE):', metrics.mean_absolute_error(y_test, p_y))
print('Mean Squared Error(MSE):', metrics.mean_squared_error(y_test, p_y))
```

```python
plt.scatter(y_test,p_y)
plt.xlabel('y_test(True Values)')
plt.ylabel('py')
plt.show()
```

```python
print('Variation Score:',metrics.explained_variance_score(y_test,p_y))
```

```python
sns.distplot((y_test-p_y),bins=50)
plt.show()
```
